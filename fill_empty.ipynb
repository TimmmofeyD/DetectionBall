```python
import os  # модуль для работы с файловой системой

# Директории, где хранятся изображения и аннотации
base_image_folder = './test/images'  # базовая директория с изображениями
base_annotation_folder = './test/labels'  # базовая директория с аннотациями

# Получаем список подкаталогов в папке с изображениями
subfolders = [f for f in os.listdir(base_image_folder) if os.path.isdir(os.path.join(base_image_folder, f))]

# Проходим по каждому подкаталогу
for subfolder in subfolders:
    # Формируем полные пути к папкам изображений и аннотаций
    image_folder = os.path.join(base_image_folder, subfolder)
    annotation_folder = os.path.join(base_annotation_folder, subfolder)
    
    # Получаем список изображений в текущем подкаталоге (файлы, оканчивающиеся на '.jpg')
    images = [f for f in os.listdir(image_folder) if f.endswith('.jpg')]
    
    # Проходим по каждому изображению
    for image in images:
        # Определяем имя файла аннотации, соответствующее изображению
        annotation_file = os.path.splitext(image)[0] + '.txt'  # заменяем расширение .jpg на .txt
        annotation_path = os.path.join(annotation_folder, annotation_file)  # полный путь к файлу аннотации
        
        # Проверяем, существует ли файл аннотации
        if not os.path.exists(annotation_path):
            # Если файл аннотации не существует, создаем пустой файл
            with open(annotation_path, 'w') as f:
                pass  # создаем пустой файл без записи в него
            # Сообщаем, что файл аннотации был создан
            print(f'создан: {annotation_path}')
```
