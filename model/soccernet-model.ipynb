```python
import cv2  # –ò–º–ø–æ—Ä—Ç –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ OpenCV –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –∏ –≤–∏–¥–µ–æ
import torch  # –ò–º–ø–æ—Ä—Ç –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ PyTorch –¥–ª—è —Ä–∞–±–æ—Ç—ã —Å —Ç–µ–Ω–∑–æ—Ä–∞–º–∏ –∏ –º–æ–¥–µ–ª—è–º–∏ –≥–ª—É–±–æ–∫–æ–≥–æ –æ–±—É—á–µ–Ω–∏—è
from ultralytics import YOLO  # –ò–º–ø–æ—Ä—Ç –∫–ª–∞—Å—Å–∞ YOLO –∏–∑ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ Ultralytics –¥–ª—è —Ä–∞–±–æ—Ç—ã —Å –º–æ–¥–µ–ª—å—é YOLO
import wandb  # –ò–º–ø–æ—Ä—Ç –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ Weights & Biases –¥–ª—è –æ—Ç—Å–ª–µ–∂–∏–≤–∞–Ω–∏—è —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–æ–≤
import os  # –ò–º–ø–æ—Ä—Ç –º–æ–¥—É–ª—è os –¥–ª—è —Ä–∞–±–æ—Ç—ã —Å —Ñ–∞–π–ª–æ–≤–æ–π —Å–∏—Å—Ç–µ–º–æ–π
import numpy as np  # –ò–º–ø–æ—Ä—Ç –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ NumPy –¥–ª—è —Ä–∞–±–æ—Ç—ã —Å –º–∞—Å—Å–∏–≤–∞–º–∏ –∏ –º–∞—Ç—Ä–∏—Ü–∞–º–∏
from collections import defaultdict  # –ò–º–ø–æ—Ä—Ç defaultdict –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è —Å–ª–æ–≤–∞—Ä–µ–π —Å –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏–º —Å–æ–∑–¥–∞–Ω–∏–µ–º –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é

# –í—Ö–æ–¥ –≤ Weights & Biases —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º –∫–ª—é—á–∞ API
wandb.login(key='#')  # –í—Å—Ç–∞–≤—å—Ç–µ —Å–≤–æ–π –∫–ª—é—á –ø—Ä–∏ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ—Å—Ç–∏ –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏ wandb
```

```python
# –°–æ–∑–¥–∞–Ω–∏–µ —ç–∫–∑–µ–º–ø–ª—è—Ä–∞ –º–æ–¥–µ–ª–∏ YOLO, –∑–∞–≥—Ä—É–∂–∞—è –ø—Ä–µ–¥–æ–±—É—á–µ–Ω–Ω—ã–µ –≤–µ—Å–∞ –∏–∑ —É–∫–∞–∑–∞–Ω–Ω–æ–≥–æ —Ñ–∞–π–ª–∞
model = YOLO('/kaggle/input/soccernet-data/yolo11n.pt')  # –ó–∞–≥—Ä—É–∑–∫–∞ –º–æ–¥–µ–ª–∏ YOLO —Å –≤–µ—Å–∞–º–∏ –∏–∑ —Ñ–∞–π–ª–∞ yolo11n.pt
```

```python
# –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ YOLO –Ω–∞ –∑–∞–¥–∞–Ω–Ω–æ–º –Ω–∞–±–æ—Ä–µ –¥–∞–Ω–Ω—ã—Ö —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º —É–∫–∞–∑–∞–Ω–Ω—ã—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
results = model.train(
    data="/kaggle/input/soccernet-data/data.yaml",  # –ü—É—Ç—å –∫ —Ñ–∞–π–ª—É —Å –∫–æ–Ω—Ñ–∏–≥—É—Ä–∞—Ü–∏–µ–π –¥–∞–Ω–Ω—ã—Ö –≤ —Ñ–æ—Ä–º–∞—Ç–µ YAML
    epochs=20,  # –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ø–æ—Ö –æ–±—É—á–µ–Ω–∏—è
    imgsz=640,  # –†–∞–∑–º–µ—Ä –≤—Ö–æ–¥–Ω—ã—Ö –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π (640x640 –ø–∏–∫—Å–µ–ª–µ–π)
    batch=256,  # –†–∞–∑–º–µ—Ä –ø–∞–∫–µ—Ç–∞ (batch size) –¥–ª—è –æ–±—É—á–µ–Ω–∏—è
    augment=True,  # –í–∫–ª—é—á–µ–Ω–∏–µ –∞—É–≥–º–µ–Ω—Ç–∞—Ü–∏–∏ –¥–∞–Ω–Ω—ã—Ö –¥–ª—è —É–≤–µ–ª–∏—á–µ–Ω–∏—è —Ä–∞–∑–Ω–æ–æ–±—Ä–∞–∑–∏—è –æ–±—É—á–∞—é—â–µ–≥–æ –Ω–∞–±–æ—Ä–∞
    fliplr=0.5,  # –í–µ—Ä—Ç–∏–∫–∞–ª—å–Ω–æ–µ (–≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω–æ–µ) –æ—Ç—Ä–∞–∂–µ–Ω–∏–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π —Å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é 50%
    scale=0.5,  # –ü—Ä–∏–º–µ–Ω–µ–Ω–∏–µ —Å–ª—É—á–∞–π–Ω–æ–≥–æ –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏—è –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –¥–æ 50%
    translate=0.1,  # –°–ª—É—á–∞–π–Ω—ã–π —Å–¥–≤–∏–≥ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –¥–æ 10% –æ—Ç –∏—Ö —Ä–∞–∑–º–µ—Ä–∞
    mixup=0.2,  # –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø—Ä–∏–º–µ–Ω–µ–Ω–∏—è —Å–º–µ—à–∏–≤–∞–Ω–∏—è –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π (mixup) –Ω–∞ —É—Ä–æ–≤–Ω–µ 20%
    mosaic=0.5,  # –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø—Ä–∏–º–µ–Ω–µ–Ω–∏—è –º–æ–∑–∞–∏—á–Ω–æ–π –∞—É–≥–º–µ–Ω—Ç–∞—Ü–∏–∏ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π (mosaic) –Ω–∞ —É—Ä–æ–≤–Ω–µ 50%
    hsv_h=0.1,  # –°–ª—É—á–∞–π–Ω–æ–µ –∏–∑–º–µ–Ω–µ–Ω–∏–µ –æ—Ç—Ç–µ–Ω–∫–∞ (hue) –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö 10%
    hsv_s=0.7,  # –°–ª—É—á–∞–π–Ω–æ–µ –∏–∑–º–µ–Ω–µ–Ω–∏–µ –Ω–∞—Å—ã—â–µ–Ω–Ω–æ—Å—Ç–∏ (saturation) –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö 70%
    hsv_v=0.4,  # –°–ª—É—á–∞–π–Ω–æ–µ –∏–∑–º–µ–Ω–µ–Ω–∏–µ —è—Ä–∫–æ—Å—Ç–∏ (value) –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö 40%
    close_mosaic=5,  # –ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –±–ª–∏–∑–∫–∏—Ö –æ–±—ä–µ–∫—Ç–æ–≤ –¥–ª—è –ø—Ä–∏–º–µ–Ω–µ–Ω–∏—è –º–æ–∑–∞–∏–∫–∏
    dropout=0.05,  # –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –æ—Ç–∫–ª—é—á–µ–Ω–∏—è (dropout) –Ω–µ–π—Ä–æ–Ω–æ–≤ –≤ 5% —Å–ª—É—á–∞–µ–≤ –≤–æ –≤—Ä–µ–º—è –æ–±—É—á–µ–Ω–∏—è
    plots=True,  # –í–∫–ª—é—á–µ–Ω–∏–µ –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤ –æ–±—É—á–µ–Ω–∏—è –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–∏
    device=[0, 1],  # –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ GPU 0 –∏ 1 –¥–ª—è —É—Å–∫–æ—Ä–µ–Ω–∏—è –æ–±—É—á–µ–Ω–∏—è
    optimizer='AdamW'  # –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ –æ–ø—Ç–∏–º–∏–∑–∞—Ç–æ—Ä–∞ AdamW –¥–ª—è –æ–±–Ω–æ–≤–ª–µ–Ω–∏—è –≤–µ—Å–æ–≤ –º–æ–¥–µ–ª–∏
)
```

```python
# –ü–∞–ø–∫–∞ —Å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è–º–∏ –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏
image_folder = '/kaggle/input/soccernet-data/challenge/SNGS-180/img1'

# –ó–∞–≥—Ä—É–∑–∫–∞ –ø—Ä–µ–¥–æ–±—É—á–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏ YOLO –∏–∑ —É–∫–∞–∑–∞–Ω–Ω–æ–≥–æ –ø—É—Ç–∏
model = YOLO("/kaggle/working/runs/detect/train/weights/best.pt")

# –ü–æ–ª—É—á–µ–Ω–∏–µ —Å–ø–∏—Å–∫–∞ —Ñ–∞–π–ª–æ–≤ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –≤ –ø–∞–ø–∫–µ, —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –∏ —Ñ–∏–ª—å—Ç—Ä–∞—Ü–∏—è –ø–æ —Ä–∞—Å—à–∏—Ä–µ–Ω–∏—é .jpg
image_files = [os.path.join(image_folder, f) for f in sorted(os.listdir(image_folder)) if f.endswith('.jpg')]

# –ù–∞—Å—Ç—Ä–æ–π–∫–∏ –¥–ª—è –≤—ã—Ö–æ–¥–Ω–æ–≥–æ –≤–∏–¥–µ–æ
output_video_path = 'output_video_180_4_25.avi'  # –ü—É—Ç—å –∫ –≤—ã—Ö–æ–¥–Ω–æ–º—É –≤–∏–¥–µ–æ
frame_width, frame_height = 1920, 1080  # –®–∏—Ä–∏–Ω–∞ –∏ –≤—ã—Å–æ—Ç–∞ –∫–∞–¥—Ä–æ–≤ –≤ –≤–∏–¥–µ–æ
fps = 25  # –ß–∞—Å—Ç–æ—Ç–∞ –∫–∞–¥—Ä–æ–≤ –≤–∏–¥–µ–æ

# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –æ–±—ä–µ–∫—Ç–∞ VideoWriter –¥–ª—è –∑–∞–ø–∏—Å–∏ –≤–∏–¥–µ–æ
fourcc = cv2.VideoWriter_fourcc(*'XVID')  # –ö–æ–¥–µ–∫ –¥–ª—è –≤–∏–¥–µ–æ
out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

# –ò—Å—Ç–æ—Ä–∏—è —Ç—Ä–µ–∫–∏–Ω–≥–∞ –¥–ª—è –æ–±—ä–µ–∫—Ç–æ–≤
track_history = defaultdict(lambda: [])  # –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ defaultdict –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è —Ç—Ä–∞–µ–∫—Ç–æ—Ä–∏–π –æ–±—ä–µ–∫—Ç–æ–≤

# –ò–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä –¥–ª—è —Ç—Ä–µ–∫–∏–Ω–≥–∞
forced_track_id = 1  
# –ü–æ—Ä–æ–≥ –¥–ª—è –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –±–æ–ª—å—à–æ–≥–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –º–µ–∂–¥—É –¥–≤—É–º—è –ø–æ–∑–∏—Ü–∏—è–º–∏
distance_threshold = 50  

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø—Ä–∏–º–µ–Ω–µ–Ω–∏—è –Ω–µ–Ω–∞–¥–ª–µ–∂–∞—â–µ–≥–æ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –ø–æ–¥–∞–≤–ª–µ–Ω–∏—è (NMS)
def apply_nms(boxes, scores, iou_threshold=0.5):
    indices = cv2.dnn.NMSBoxes(boxes.tolist(), scores.tolist(), score_threshold=0.1, nms_threshold=iou_threshold)
    if len(indices) > 0:
        indices = indices.flatten()  # –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –∏–Ω–¥–µ–∫—Å–æ–≤ –≤ –æ–¥–Ω–æ–º–µ—Ä–Ω—ã–π –º–∞—Å—Å–∏–≤
    return indices

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –º–µ–∂–¥—É –¥–≤—É–º—è —Ç–æ—á–∫–∞–º–∏
def calculate_distance(p1, p2):
    return np.linalg.norm(np.array(p1) - np.array(p2))  # –ò—Å–ø–æ–ª—å–∑—É–µ–º –Ω–æ—Ä–º—É –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –µ–≤–∫–ª–∏–¥–æ–≤–∞ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è

# –ü—Ä–æ—Ö–æ–¥ –ø–æ –≤—Å–µ–º –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è–º –≤ –ø–∞–ø–∫–µ
for image_file in image_files:
    frame = cv2.imread(image_file)  # –ß—Ç–µ–Ω–∏–µ —Ç–µ–∫—É—â–µ–≥–æ –∫–∞–¥—Ä–∞

    # –û—Ç—Å–ª–µ–∂–∏–≤–∞–Ω–∏–µ –æ–±—ä–µ–∫—Ç–æ–≤ –Ω–∞ —Ç–µ–∫—É—â–µ–º –∫–∞–¥—Ä–µ —Å –ø–æ–º–æ—â—å—é –º–æ–¥–µ–ª–∏ YOLO
    results = model.track(frame, persist=True, conf=0.1)

    # –û—Ç—Ä–∏—Å–æ–≤–∫–∞ –∞–Ω–Ω–æ—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –∫–∞–¥—Ä–∞
    annotated_frame = results[0].plot()  
    boxes = results[0].boxes.xywh.cpu().numpy()  # –ü–æ–ª—É—á–µ–Ω–∏–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç –∏ —Ä–∞–∑–º–µ—Ä–æ–≤ bounding box (x_center, y_center, width, height)
    scores = results[0].boxes.conf.cpu().numpy()  # –ü–æ–ª—É—á–µ–Ω–∏–µ —É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç–∏ (confidence scores) –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–Ω–æ–≥–æ –æ–±—ä–µ–∫—Ç–∞
    track_ids = results[0].boxes.id  # –ü–æ–ª—É—á–µ–Ω–∏–µ –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä–æ–≤ –æ–±—ä–µ–∫—Ç–æ–≤

    # –ü—Ä–∏–º–µ–Ω—è–µ–º NMS –¥–ª—è —É–¥–∞–ª–µ–Ω–∏—è –ø–µ—Ä–µ–∫—Ä—ã–≤–∞—é—â–∏—Ö—Å—è bounding box
    if len(boxes) > 0:
        # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –∏–∑ —Ñ–æ—Ä–º–∞—Ç–∞ xywh –≤ xyxy
        boxes_xyxy = np.hstack([boxes[:, :2] - boxes[:, 2:4] / 2, boxes[:, :2] + boxes[:, 2:4] / 2])  
        indices = apply_nms(boxes_xyxy, scores)

        if len(indices) > 0:
            for i in indices:
                box = boxes[i]  # –ü–æ–ª—É—á–µ–Ω–∏–µ —Ç–µ–∫—É—â–µ–≥–æ bounding box
                
                if track_ids is None or len(track_ids) == 0 or track_ids[i] is None:
                    track_id = forced_track_id
                else:
                    track_id = track_ids[i].int().cpu().item()  # –ü–æ–ª—É—á–µ–Ω–∏–µ –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä–∞ –æ–±—ä–µ–∫—Ç–∞

                # –ò–∑–≤–ª–µ—á–µ–Ω–∏–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç —Ü–µ–Ω—Ç—Ä–∞ bounding box
                x_center, y_center, _, _ = box

                track = track_history[track_id]  # –ü–æ–ª—É—á–µ–Ω–∏–µ –∏—Å—Ç–æ—Ä–∏–∏ —Ç—Ä–µ–∫–∏–Ω–≥–∞ –¥–ª—è —Ç–µ–∫—É—â–µ–≥–æ –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä–∞
                if len(track) > 0:  # –ï—Å–ª–∏ –µ—Å—Ç—å –ø—Ä–µ–¥—ã–¥—É—â–∏–µ –ø–æ–∑–∏—Ü–∏–∏
                    last_position = track[-1]  # –ü–æ—Å–ª–µ–¥–Ω—è—è –∏–∑–≤–µ—Å—Ç–Ω–∞—è –ø–æ–∑–∏—Ü–∏—è
                    distance = calculate_distance(last_position, (x_center, y_center))  # –í—ã—á–∏—Å–ª–µ–Ω–∏–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –¥–æ —Ç–µ–∫—É—â–µ–π –ø–æ–∑–∏—Ü–∏–∏
                    
                    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ —Å–ª–∏—à–∫–æ–º –±–æ–ª—å—à–æ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ (—Å–∫–∞—á–æ–∫)
                    if distance > distance_threshold:
                        print(f"–û–±—ä–µ–∫—Ç {track_id} –ø—Ä–æ–ø—É—â–µ–Ω –∏–∑-–∑–∞ —Å–∫–∞—á–∫–∞: —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ {distance:.2f}")
                        continue  # –ü—Ä–æ–ø—É—Å–∫–∞–µ–º —Ç—Ä–µ–∫–∏–Ω–≥ –¥–ª—è —ç—Ç–æ–≥–æ –æ–±—ä–µ–∫—Ç–∞

                # –î–æ–±–∞–≤–ª–µ–Ω–∏–µ —Ç–µ–∫—É—â–µ–π –ø–æ–∑–∏—Ü–∏–∏ –≤ –∏—Å—Ç–æ—Ä–∏—é —Ç—Ä–µ–∫–∏–Ω–≥–∞
                track.append((float(x_center), float(y_center)))

                # –£–¥–∞–ª–µ–Ω–∏–µ —Å—Ç–∞—Ä—ã—Ö –ø–æ–∑–∏—Ü–∏–π, –µ—Å–ª–∏ –∏—Ö –±–æ–ª—å—à–µ 10
                if len(track) > 10:  
                    track.pop(0)

                # –û—Ç—Ä–∏—Å–æ–≤–∫–∞ —Ç—Ä–∞–µ–∫—Ç–æ—Ä–∏–∏ –¥–≤–∏–∂–µ–Ω–∏—è –æ–±—ä–µ–∫—Ç–∞
                points = np.array(track).astype(np.int32).reshape((-1, 1, 2))  # –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –≤ –Ω—É–∂–Ω—ã–π —Ñ–æ—Ä–º–∞—Ç
                cv2.polylines(annotated_frame, [points], isClosed=False, color=(230, 230, 230), thickness=2)  # –û—Ç—Ä–∏—Å–æ–≤–∫–∞ –ø–æ–ª–∏–ª–∏–Ω–∏–π

    # –ò–∑–º–µ–Ω–µ–Ω–∏–µ —Ä–∞–∑–º–µ—Ä–∞ –∞–Ω–Ω–æ—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –∫–∞–¥—Ä–∞ –¥–æ –∑–∞–¥–∞–Ω–Ω—ã—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
    annotated_frame = cv2.resize(annotated_frame, (frame_width, frame_height))

    # –ó–∞–ø–∏—Å—å –∞–Ω–Ω–æ—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –∫–∞–¥—Ä–∞ –≤ –≤—ã—Ö–æ–¥–Ω–æ–µ –≤–∏–¥–µ–æ
    out.write(annotated_frame)

# –û—Å–≤–æ–±–æ–∂–¥–µ–Ω–∏–µ —Ä–µ—Å—É—Ä—Å–æ–≤, —Å–≤—è–∑–∞–Ω–Ω—ã—Ö —Å –∑–∞–ø–∏—Å—å—é –≤–∏–¥–µ–æ
out.release()

# –°–æ–æ–±—â–µ–Ω–∏–µ –æ –∑–∞–≤–µ—Ä—à–µ–Ω–∏–∏ —Ä–∞–±–æ—Ç—ã –∏ —É–∫–∞–∑–∞–Ω–∏–µ –ø—É—Ç–∏ –∫ –≤—ã—Ö–æ–¥–Ω–æ–º—É –≤–∏–¥–µ–æ
print(f"–≤–∏–¥–µ–æ {output_video_path}")
```
